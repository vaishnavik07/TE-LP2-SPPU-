Group by categorical variable gender

```python
df.info()
```

```python
columns=['Age','Salary','Experience_Years']

for column in columns:
    print('{:<20}{:<10}{:<10}{:<20}{:<10}{:10}'.format('columns','min','max','mean','median','std'))
    m1=df[column].groupby(df['Gender']).min()
    m2=df[column].groupby(df['Gender']).max()
    m3=df[column].groupby(df['Gender']).mean()
    m4=df[column].groupby(df['Gender']).median()
    m5=df[column].groupby(df['Gender']).std()
    print('{:<20}{:<10}{:<10}{:<20}{:<10}{:10}'.format("Female",m1[0],m2[0],m3[0],m4[0],m5[0]))
    print('{:<20}{:<10}{:<10}{:<20}{:<10}{:10}'.format("Male",m1[1],m2[1],m3[1],m4[1],m5[1]))
```
