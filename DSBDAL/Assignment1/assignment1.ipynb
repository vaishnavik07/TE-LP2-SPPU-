Handle outliers using interquartile range

```python
Q1= df['writing score'].quantile(0.25)
Q3= df['writing score'].quantile(0.75)

IQR=Q3-Q1

print(Q1)
print(Q3)
print(IQR)
```

```python
outlier_threshold=1.5*IQR
print(outlier_threshold)
```

```python
# Identify the outliers
outliers=df[(df['writing score']<Q1-outlier_threshold) | (df['writing score']>Q3+outlier_threshold)]
print(outliers)
```

```python
#Replace oulier values by boundary values  
df_transformed=df.copy()

df_transformed['writing score']=np.clip(df_transformed['writing score'], Q1-outlier_threshold, Q3+outlier_threshold)
```

```python
outliers_rem=df_transformed[(df_transformed['writing score']<Q1-outlier_threshold) | (df_transformed['writing score']>Q3+outlier_threshold)]
outliers_rem
```

```python
#Option 2: remove outliers

df_filtered=df[~((df['writing score']<Q1-outlier_threshold) | (df['writing score']>Q3+outlier_threshold))]
df_filtered
```
